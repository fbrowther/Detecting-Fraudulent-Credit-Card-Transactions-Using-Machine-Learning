```python

file_path = Path("./Resources/creditcardfraud_normalised.csv")
df = pd.read_csv(file_path)
df
```

```python
from collections import Counter
#
target = df.values[:, -1]
counter = Counter(target)
for k,v in counter.items():
  per = v/len(target)*100
  print("Class=%d, Count=%d, Percentage=%.3f%%" % (k,v, per))
```

```python
# Remove class target from features data
Labels = df["class"].values
Features = df.drop(columns="class").values
```

```python
mergings = linkage(Features, method="ward")
```

```python
plt.figure()
dendrogram(mergings, leaf_rotation=90, leaf_font_size=5)
plt.ylabel('Euclidean Distance')
plt.title('Cluster Dendrogram')
plt.show()
```
